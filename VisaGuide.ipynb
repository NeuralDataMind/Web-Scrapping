```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import requests
```

```python
density_response = requests.get("https://visaguide.world/asia/")
```

```python
density_response.text
```

```python
type(density_response.text)
```

```python
densitysoup=BeautifulSoup(density_response.text,"html.parser")
```

```python
type(densitysoup)
```

```python
density_tables=densitysoup.find_all("table")
```

```python
len(density_tables)
```

```python
density_headers_tag = density_tables[1].find_all("th")
print(density_headers_tag)
```

```python
density_headers=[i.text for i in density_headers_tag ]
print(density_headers)
```

```python
density_rows_tag=density_tables[1].find_all("td")
print(density_rows_tag)
```

```python
density_rows=[i.text for i in density_rows_tag ]
print(density_rows)
```

```python
density_dict={}
n=0
for i in density_headers:
    density_dict[i]=[density_rows[j] for j in range(n,len(density_rows),len(density_headers)) ]
    n+=1
density_dict
```

```python
density_df=pd.DataFrame(density_dict)
density_df
```

```python
density_df.index=range(1,len(density_df)+1)
density_df
```

```python
density_df.to_csv("density_data.csv")
```

```python
density_df["Population (2021)"] = density_df["Population (2021)"].str.replace(",","")
density_df["Population (2021)"] = density_df["Population (2021)"].astype(int)
density_df
```

```python
density_df["Area km2"] = density_df["Area km2"].str.replace(",","")
density_df["Area km2"] = density_df["Area km2"].astype(int)
density_df
```

```python
density_df["population_density"] = density_df["Population (2021)"]/density_df['Area km2']
density_df
```

```python
import matplotlib.pyplot as plt
x=density_df["Country"][:15]
y=density_df["population_density"][:15]
plt.xticks(rotation=90)
plt.bar(x,y)
plt.xlabel("Country")
plt.ylabel("population_density")
plt.title("population_density")
plt.show()
```
